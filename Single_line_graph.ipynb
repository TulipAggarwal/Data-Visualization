```python
#First graph-Single line graph
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[1,2,3,4]
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,y,color='pink', linestyle='dotted', marker='o', markerfacecolor='black' , markersize= 12, linewidth=3)
plt.title("first graph")
plt.show()
```
