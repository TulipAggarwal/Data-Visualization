```python
#Bar graph
import matplotlib.pyplot as plt
import numpy as np
y=[1,2,3,4]
x=[5,6,7,8]
tick_label=['one', 'two','three' ,'four']
plt.bar(np.arange(len(y))-0.2, y, tick_label= tick_label, width=0.4)
plt.bar(np.arange(len(x))+0.2, x, tick_label= tick_label, width=0.4)
plt.title("my bar chart")
plt.legend(labels=['a','b'])
plt.show()
```

```python
#Histogram
import matplotlib.pyplot as plt
ages=[10,20,30,20,40,44,45,50,60,70,20,21,44,9]
range=(0,100)
bins=10
plt.hist(ages,bins,range,width=10,edgecolor='black')
plt.xlabel("age")
plt.ylabel("number of people")
plt.title("my histogram")
plt.show()
```

```python
#Pie chart
import matplotlib.pyplot as plt
activities=['eat','sleep','work','play']
slices=[3,7,8,6]
colors=['r','b','y','g']
plt.pie(slices,labels=activities,colors=colors)
plt.title("My pie chart")
plt.show()
```
