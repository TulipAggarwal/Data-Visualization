```python
# 3-D Plot 
import numpy as np 
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d 
plt.axes(projection='3d') 
x=[1,2,3,4] 
y=[5,6,7,8] 
z=[4,5,6,6] 
plt.xlabel("X axis") 
plt.ylabel("Y axis") 
plt.plot(x,y,z) 
plt.show() 
```

```python
#Contour plot 
import matplotlib.pyplot as plt 
import numpy as np 
feature_x=np.linspace(0,3.0,30) 
feature_y=np.linspace(0,3.0,30) 
[X,Y]= np.meshgrid(feature_x,feature_y) 
Z=X**2+Y**2 
plt.contourf(X,Y,Z) 
plt.show() 
```
