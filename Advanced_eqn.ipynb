```python
#Free fall 
import numpy as np 
import matplotlib.pyplot as plt 
m = 10 
d = 20 
g = 9.81 
t_max = (2*d/g)**0.5 
t=np.arange(0,t_max,0.01) 
h = d-(0.5*(g*t*t)) 
v = g*t 
ke = 0.5*v*v*m 
pe = m*g*h 
plt.plot(t,h,'r', label='Height') 
plt.plot(t,v,'b', label='velocity') 
plt.plot(t,ke,'y', label='kinetic energy') 
plt.plot(t,pe,'k', label='potential energy') 
plt.legend() 
plt.show()
```

```python
#Projectile in 2-D 
import numpy as np 
import matplotlib.pylab as plot 
import math as m 
v = 30 
g = 9.8 
theta = np.arange(m.pi/6, m.pi/3, m.pi/36) 
t = np.linspace(0, 5, num=100) 
for i in theta: 
 x1 = [] 
 y1 = [] 
for k in t: 
 x = ((v*k)*np.cos(i)) 
 y = ((v*k)*np.sin(i))-((0.5*g)*(k**2)) 
 x1.append(x) 
 y1.append(y) 
 p = [i for i, j in enumerate(y1) if j < 0] 
for i in sorted(p, reverse = True): 
 del x1[i] 
 del y1[i] 
 plot.plot(x1, y1) 
plot.show()
```
