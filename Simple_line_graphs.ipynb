```python
#First graph-Single line graph
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[1,2,3,4]
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,y,color='pink', linestyle='dotted', marker='o', markerfacecolor='black' , markersize= 12, linewidth=3)
plt.title("first graph")
plt.show()
```

```python
#Two lines on same graph
import matplotlib.pyplot as plt
x1=[1,3,5,7]
y1=[2,4,6,9]
plt.plot(x1,y1,marker='o', color='magenta' , linestyle = 'dotted' , markerfacecolor='black')
x2=[1,2,3,4]
y2=[4,3,2,1]
plt.plot(x2,y2, marker='s' , color='yellow' , linestyle='dashed' , markerfacecolor='red')
plt.xlabel("x axis", fontsize=16)
plt.ylabel("y axis", fontsize=16)
plt.title("Two lines on same graph", fontsize=20)
plt.show()
```
